```python
import requests
from bs4 import BeautifulSoup

# 웹 페이지 URL 설정
url = '여기에 웹 페이지 주소를 입력하세요'

# requests를 이용해 웹 페이지 가져오기
response = requests.get(url)

# BeautifulSoup 객체 생성
soup = BeautifulSoup(response.text, 'html.parser')

# 두 번째 테이블 선택
second_table = soup.find_all('table')[1]

# 두 번째 테이블에서 두 번째 'td' 내의 두 번째 'td' 선택
second_td_in_second_table = second_table.find_all('td')[1].find_all('td')[1]

# 위에서 선택된 td 내의 네 번째 테이블 ('a'로 지정)
table_a = second_td_in_second_table.find_all('table')[3]

# 'a' 테이블의 모든 행 ('tr')을 순환
data = []
for tr in table_a.find_all('tr'):
    # 각 행에서 세 번째 'td' 선택
    third_td = tr.find_all('td')[2]
    
    # 세 번째 'td' 내의 두 번째 테이블 선택
    second_table_in_third_td = third_td.find_all('table')[1]
    
    # 특정 행 (2, 4, 7번 행)의 특정 'td' (3번째, 6번째) 값을 가져오기
    rows_of_interest = [1, 3, 6]  # 인덱스는 0부터 시작하므로 1 빼줍니다
    cols_of_interest = [2, 5]    # 마찬가지로 인덱스는 0부터 시작
    
    for row_index in rows_of_interest:
        row = second_table_in_third_td.find_all('tr')[row_index]
        for col_index in cols_of_interest:
            value = row.find_all('td')[col_index].text.strip()
            data.append(value)

# 결과 확인
print(data)
```

```python
import requests
from bs4 import BeautifulSoup

# 웹 페이지 URL 설정
url = '여기에 웹 페이지 주소를 입력하세요'

# requests를 이용해 웹 페이지 가져오기
response = requests.get(url)

# BeautifulSoup 객체 생성
soup = BeautifulSoup(response.text, 'html.parser')

# 테이블 찾기 (첫 번째 테이블을 예로 들겠습니다)
table = soup.find('table')

# 테이블의 모든 행 찾기
rows = table.find_all('tr')

# 이차원 배열로 테이블 데이터 저장
table_data = []
for row in rows:
    cols = row.find_all('td')  # 테이블 데이터 셀(td) 찾기
    cols = [ele.text.strip() for ele in cols]  # 각 셀의 텍스트 데이터 추출 및 정리
    table_data.append(cols)

# 결과 확인
for row in table_data:
    print(row)
```
